```python
import re
import random
import time
import math
from collections import Counter, defaultdict
import matplotlib.pyplot as plt

text = open("dataset.txt", encoding="utf-8").read().lower()
text = re.sub(r"[^a-z\s]", " ", text)
words = text.split()

vocab = Counter(words)
total_words = sum(vocab.values())

bigrams = defaultdict(Counter)
trigrams = defaultdict(Counter)

for i in range(len(words) - 1):
    bigrams[words[i]][words[i+1]] += 1

for i in range(len(words) - 2):
    trigrams[(words[i], words[i+1])][words[i+2]] += 1

def autocomplete_prefix(prefix, k=5):
    matches = [w for w in vocab if w.startswith(prefix)]
    matches.sort(key=lambda x: vocab[x], reverse=True)
    return matches[:k]

def autocomplete_next(prev_word, k=5):
    if prev_word in bigrams:
        return [w for w, _ in bigrams[prev_word].most_common(k)]
    return []

alphabet = "abcdefghijklmnopqrstuvwxyz"

def edits1(word):
    splits = [(word[:i], word[i:]) for i in range(len(word)+1)]
    deletes = [L + R[1:] for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]
    replaces = [L + c + R[1:] for L, R in splits if R for c in alphabet]
    inserts = [L + c + R for L, R in splits for c in alphabet]
    return set(deletes + transposes + replaces + inserts)

def autocorrect(word):
    candidates = edits1(word)
    known = [w for w in candidates if w in vocab]
    if not known:
        return word
    return max(known, key=lambda w: vocab[w])

def evaluate_autocorrect(samples=100):
    correct = 0
    test_words = random.sample(list(vocab.keys()), samples)
    for w in test_words:
        if len(w) < 4:
            continue
        error = w[:-1]
        if autocorrect(error) == w:
            correct += 1
    return correct / samples

def evaluate_autocomplete(samples=100):
    correct = 0
    for _ in range(samples):
        i = random.randint(0, len(words) - 2)
        prev = words[i]
        actual = words[i+1]
        preds = autocomplete_next(prev)
        if actual in preds:
            correct += 1
    return correct / samples

ac_accuracy = evaluate_autocomplete()
au_accuracy = evaluate_autocorrect()

labels = ["Autocomplete", "Autocorrect"]
values = [ac_accuracy, au_accuracy]

plt.bar(labels, values)
plt.ylim(0, 1)
plt.title("Model Accuracy Comparison")
plt.show()

print("Autocomplete Accuracy:", ac_accuracy)
print("Autocorrect Accuracy:", au_accuracy)

print("Prefix Autocomplete:", autocomplete_prefix("th"))
print("Next Word Prediction:", autocomplete_next("the"))
print("Autocorrect Example:", autocorrect("recieve"))
```
